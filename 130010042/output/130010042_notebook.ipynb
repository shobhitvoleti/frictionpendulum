# Friction Pendulum 
* Observing the response of a system considering effects of friction due to air surrounding the bob of the pendulum
* Computation achieved with the aid of Scipy's Integrate function
* Animation achieved using MATPLOTLIB's animation tool
* Plots achieved using MATPLOTLIB's Pyplot

```python
import matplotlib.animation as animation
from scipy.integrate import odeint
from numpy import arange
from pylab import *
import matplotlib.pyplot as plt
from matplotlib import animation, rc
%matplotlib inline
#%matplotlib osx
from IPython.display import HTML

tnrfont = {'fontname':'Times New Roman'}

from IPython.display import HTML


#Define the system of ODEs to be solved
def FrictionPendulum(state, t):
    vel, ang = state
    lam = 0.5
    d_vel = -(lam * vel + 9.8 * ang)
    d_ang = vel
    return [d_vel, d_ang]

#Generate array for time and define initial state
t = arange(0, 20, 0.1)
init_state = [1, 1]

#Solving ODE system using odeint function
state = odeint(FrictionPendulum, init_state, t)

#Create figure for animation
fig = plt.figure()
ax = plt.axes(xlim=(-5, 5), ylim=(-5, 5))
line, = ax.plot([], [], lw=2)

#Plot the solution obtained from above odeint
plt.xlabel("Angular Position",fontsize = 12,**tnrfont)
plt.ylabel("Angular Velocity",fontsize = 12,**tnrfont)

def init():
    line.set_data([], [])
    return line,


#Function to animate plot
def animate(i):
    line.set_data(state[0:i, 0], state[0:i, 1])
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=200, interval=20, blit=True)
plt.grid()

plt.close()
HTML(anim.to_html5_video())
```
